```python
# In the PIAAC data sets, there are 2 population columns. The POP_DOMAIN coloumn is the column we should be using. 
df_county_prepared = df_county_dropped.drop(columns=['POP'])
```

```python
# Rechecking County data
df_county_prepared.isnull().sum()
```

```python
df_county_prepared.dtypes
```

#### The county (along with the state and nation) data set has a large number of missing values. As this project is for exploratory purposes, I will impute the missing values using that column's mean. This means that there will be bias. As there isn't a lot of available data to work off of, there will be numerous repeats of the same values which will cause some odd grouping in graphs. Since we are imputing a large chunk of the data, this data is unable to tell us how counties in the United States really differ.  

#### Alternatively, instead of imputing with the mean, I could build a regression to predict the imputed value, that way there would be variation in the value dependent on the other values in the other data set.

```python
# Imputing missing values in county data set
df_county_prepared[['Male', 'Female', 'White', 'Black', 'Hispanic', 'Asian', 'AIAN', 'NHPI', 'Other_race', 'Less_HS', 'HS',
                   'More_HS', 'Eng_not_well', 'FB_after2010', 'FB_1990_2009', 'FB_before1990', 'FB', 'Poverty_100', 'Poverty_150', 'SNAP',
                    'Employed', 'Unemployed', 'Not_in_labor', 'OCC_Manage', 'OCC_Service', 'OCC_Sales', 'OCC_Natural', 'OCC_Military', 'OCC_Prod',
                    'No_Insurance']] = (
    df_county_prepared[['Male', 'Female', 'White', 'Black', 'Hispanic', 'Asian', 'AIAN', 'NHPI', 'Other_race', 'Less_HS', 'HS',
                   'More_HS', 'Eng_not_well', 'FB_after2010', 'FB_1990_2009', 'FB_before1990', 'FB', 'Poverty_100', 'Poverty_150', 'SNAP',
                    'Employed', 'Unemployed', 'Not_in_labor', 'OCC_Manage', 'OCC_Service', 'OCC_Sales', 'OCC_Natural', 'OCC_Military', 'OCC_Prod',
                    'No_Insurance']]
    .apply(lambda col: col.fillna(col.mean()))
)
```
