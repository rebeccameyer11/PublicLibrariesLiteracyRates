# 04. Correlation Matrix Heatmap

```python
# Selecting columns to use for heatmap

combined_1 = pd.concat([df_public_libraries['LOCGVT'], df_public_libraries['STGVT'], df_public_libraries['FEDGVT'], df_public_libraries['OTHINCM'], 
                    df_county['Lit_P1'], df_county['Lit_P2'], df_county['Lit_P3'], df_county['Lit_A']], axis=1)
```

```python
# Creating the correlation heatmap using pandas

combined_1.corr()
```

```python
# Creating a correlation heatmap using matplotlib

plt.matshow(combined_1.corr())
plt.show()
```

```python
# Save figure
plt.matshow(combined_1.corr())
plt.savefig("heatmap_1.png") 
```

```python
#current dir
cwd = os.getcwd()
cwd
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(combined_1.corr(), cmap='plasma')
plt.xticks(range(combined_1.shape[1]), combined_1.columns, fontsize=10, rotation=45)
plt.yticks(range(combined_1.shape[1]), combined_1.columns, fontsize=10)
plt.colorbar()
plt.title('Correlation Heatmap of Library Funding and Literacy Rates', fontsize=14) # add title
plt.show()
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(combined_1.corr(), annot = True, ax = ax)

# Save figure
plt.savefig("heatmap_2.png") 
```

### Correlation Matrix Heatmap Description

#### Looking at the correlation heatmap and the values of the separate data sets, all correlations are basically at zero. That means that there is no correlation between any kind of government funding (local, state, or federal) and literacy rates. While some values are slightly larger than others, it's still not enough to determine or distinguish any correlation. These two data sets may not be the best for using as a heatmap, but it does show that funding isn't a direct literacy driver. Even if the library receives more local government money, that funding may go elsewhere (facilities, staffing) or to programs not related to literacy. These data sets may also be mismatched - funding happens only once a year and may not reflect long-term patterns.
